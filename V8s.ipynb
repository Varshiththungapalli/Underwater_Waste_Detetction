Evaluation

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from ultralytics import YOLO
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

# Load trained YOLO model
model = YOLO(r"C:\Users\VARSHITH\OneDrive - WOXSEN UNIVERSITY\YEAR-3 SEM 5&6\Applicative Scratch\YoloV8_RI\V8s\trained_model.pt")  

# Define paths
test_images_path = r"C:\Users\VARSHITH\OneDrive - WOXSEN UNIVERSITY\YEAR-3 SEM 5&6\Applicative Scratch\RI\test\images"  
test_labels_path = r"C:\Users\VARSHITH\OneDrive - WOXSEN UNIVERSITY\YEAR-3 SEM 5&6\Applicative Scratch\RI\test\labels"  

# Lists to hold true and predicted labels
true_labels = []
pred_labels = []

# Run inference
results = model(test_images_path)

# Extract true labels and predictions
for result in results:
    image_filename = os.path.splitext(os.path.basename(result.path))[0]
    label_file_path = os.path.join(test_labels_path, f"{image_filename}.txt")

    if os.path.exists(label_file_path):
        with open(label_file_path, 'r') as f:
            true_image_labels = [int(line.split()[0]) for line in f.readlines()]
        
        num_pred_labels = len(result.boxes.cls) if result.boxes.cls is not None else 0

        if len(true_image_labels) == num_pred_labels:
            true_labels.extend(true_image_labels)
            pred_labels.extend(result.boxes.cls.cpu().numpy())
        else:
            min_labels = min(len(true_image_labels), num_pred_labels)
            true_labels.extend(true_image_labels[:min_labels])
            pred_labels.extend(result.boxes.cls.cpu().numpy()[:min_labels])

# Convert to numpy arrays
true_labels = np.array(true_labels)
pred_labels = np.array(pred_labels)

# Ensure valid labels before computing metrics
if len(true_labels) == 0 or len(pred_labels) == 0:
    print("Error: Labels are empty. Check file paths and label extraction.")
elif len(true_labels) != len(pred_labels):
    print(f"Error: Mismatch in true and predicted labels. True: {len(true_labels)}, Predicted: {len(pred_labels)}")
else:
    # Compute Metrics
    accuracy = accuracy_score(true_labels, pred_labels)
    precision = precision_score(true_labels, pred_labels, average='weighted', zero_division=1)
    recall = recall_score(true_labels, pred_labels, average='weighted', zero_division=1)
    f1 = f1_score(true_labels, pred_labels, average='weighted', zero_division=1)
    conf_matrix = confusion_matrix(true_labels, pred_labels)

    # Print Metrics
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1 Score: {f1:.4f}")

    # Plot Confusion Matrix
    plt.figure(figsize=(8, 6))
    sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.title("Confusion Matrix")
    plt.show()

    # True vs Predicted Graph
    plt.figure(figsize=(8, 6))
    plt.hist(true_labels, bins=np.arange(max(true_labels)+2)-0.5, alpha=0.6, label="True Labels", color='blue')
    plt.hist(pred_labels, bins=np.arange(max(pred_labels)+2)-0.5, alpha=0.6, label="Predicted Labels", color='orange')
    plt.xlabel("Class Labels")
    plt.ylabel("Frequency")
    plt.title("True vs Predicted Labels Distribution")
    plt.legend()
    plt.show()
```
